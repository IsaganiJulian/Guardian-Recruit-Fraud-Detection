### Master Data Split
- The df was split using a stratified-y, which was the target variable being 'Fraudeulent'
- This is necessary since there is an imbalance between the fraudelent listings and the legitimate ones. 

```python
from sklearn.model_selection import train_test_split

# Define your target variable 
target_col = 'fraudulent' 

# First split: 70% Train, 30% Remainder [Stratified]
train_df, rem_df = train_test_split(
    df, 
    train_size=0.7, 
    random_state=42, 
    stratify=df[target_col]  # <--- Project-critical change
)

# Second split: Split remainder equally (15% Val, 15% Test) [Stratified]
val_df, test_df = train_test_split(
    rem_df, 
    test_size=0.5, 
    random_state=42, 
    stratify=rem_df[target_col] # <--- Stratify based on the remainder's labels
)

print(f"Train: {len(train_df)} | Val: {len(val_df)} | Test: {len(test_df)}")
```

```python
# Save train, test, val to .csv
train_df.to_csv('../data/processed/train.csv', index=False)
val_df.to_csv('../data/processed/val.csv', index=False)
test_df.to_csv('../data/processed/test.csv', index=False)

print("Files saved to data/processed/")
```
